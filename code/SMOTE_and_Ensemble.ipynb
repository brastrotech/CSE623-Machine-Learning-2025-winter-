# With SMOTE

# T_eff

```python
file_name = 'data_output_teff.pkl'
n_features = 9

X_train, X_test, y_train, y_test = u.load_pkl_data(file_name)
X_train, X_test = X_train[:,:n_features], X_test[:,:n_features]

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# Initialize SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)

# Apply SMOTE only on the training set
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Check new class distribution
print("Class distribution after SMOTE:", Counter(y_train_resampled))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_resampled = scaler.fit_transform(X_train_resampled)
X_test = scaler.transform(X_test)  # Scale test data using same scaler
```

```python
# Initialize KNN model
knn = KNeighborsClassifier(n_neighbors=11)  # You can tune 'n_neighbors'

# Train the model
knn.fit(X_train_resampled, y_train_resampled)
```

```python
# Predict on test set
y_pred = knn.predict(X_test)

# Evaluate performance
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred,labels=[0,1,2,3,4,5,6])

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["M", "K", "G", "F", "A", "B", "O"])

disp.plot(cmap=plt.cm.Blues)
plt.title('$T_{eff}$ (After SMOTE)', fontsize=15, pad=20)
plt.xlabel('Prediction', fontsize=11)
plt.ylabel('Actual', fontsize=11)
plt.gca().xaxis.set_label_position('top')
plt.gca().xaxis.tick_top()
# plt.gca().figure.subplots_adjust(bottom=0.2)
plt.gca().figure.text(0.5, 0.05, 'Prediction', ha='center', fontsize=13)

plt.savefig('t_eff_SMOTE.png',bbox_inches='tight',dpi=600)
plt.show()
```
